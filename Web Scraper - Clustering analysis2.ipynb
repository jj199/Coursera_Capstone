# Wikipedia Webscraper - Data analysis - neighborhoods in Toronto

First I imported BeautifulSoup and requests. 
I then sent a request to fetch the html information from the wikipedia page and saved it under the variable 'source'.

The source variable was then parsed using Beautiful Soup 4 and lxml package.

```python
from bs4 import BeautifulSoup
import requests

source = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M ').text
soup = BeautifulSoup(source, 'lxml')
```

I created a table variabe that only fetched the 'tbody' tags in the html file as this pertained to the table html code.

Further, the links variable was created to find all the 'th' tags which are the table headings - this creates a list with the tags included. 

I ran a loop through the list to only get the text and remove the tags and new lines and appended to new list called 'columns'.

These are going to be the columns for the dataframe.

```python
table = soup.find('tbody')

links = table.findAll('th')
columns = []
for link in links:
    link = link.get_text().strip('\n')
    columns.append(link)
print(columns) 
```

The 'contents' variable was created, where similar to above, all the 'td' tags were fetched from the table html file.

Same as above, the contents variable is a list with tags included, so I used a for loop to go through each item in the list and get the text/remove the tags and strip any new lines.

The result is a list of all the items in the table.

```python
contents = table.findAll('td')
#print(contents)

contentslist = []
for cont in contents:
    cont = cont.get_text().strip('\n')
    contentslist.append(cont)
print(contentslist)
```

The above list had to be separated into the corresponding separate lists for each heading. 

It is clear that first item and every 3rd item after that were all the postcodes.
Similarly, the second item and every 3rd item after that were all Boroughs and every 3rd item and 3rd item after that are the Neighborhoods.

As such new lists were created using list index/slicing for each of those columns in preparation for inserting into dataframe.
Below output is 3 separate lists as specified.

```python
postcodes = contentslist[0::3]
boroughs = contentslist[1::3]
neighborhoods = contentslist[2::3]

print(postcodes)
print(boroughs)
print(neighborhoods)

    
```

In order to create new dataframe, pandas was imported as pd and new dataframe was created.

The relevant data for the columns were inserted with the corresponding lists.

The dataframe cleaned - all the rows were the borough was not assigned was removed. 

The dataframe was grouped by Postcode, and the neighborhoods in those neighborhoods were joined together in the same row being separated by commas (,). The index was then reset.


Furthermore, I used iloc to locate all the rows where the neigborhood is "Not assigned" and this was replaced with the Borough of that same row. 
The change can be seen in row 85

```python
import pandas as pd 
df = pd.DataFrame()
df['Postcode'] = postcodes
df['Borough'] = boroughs
df['Neighborhood'] = neighborhoods

df = df[df.Borough !='Not assigned']


df = df.groupby(['Postcode','Borough'])['Neighborhood'].apply(', '.join).reset_index()

df.loc[df['Neighborhood'] == 'Not assigned','Neighborhood'] = df['Borough']
df
```

The shape of the dataframe was then obtained as requested.

```python
df = pd.concat([df, df1[['Latitude','Longitude']]], axis=1)

df
```

```python
address = 'Toronto, Ontaio'

geolocator = Nominatim(user_agent="ny_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print(location)
print('The geograpical coordinate of Toronto are {}, {}.'.format(latitude, longitude))
```

```python
map_toronto = folium.Map(location=[latitude, longitude], zoom_start=10)

# add markers to map
for lat, lng, borough, neighborhood in zip(df['Latitude'], df['Longitude'], df['Borough'], df['Neighborhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```
